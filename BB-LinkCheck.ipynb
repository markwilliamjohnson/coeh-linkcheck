```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from PIL import Image
import Screenshot
import time
import pandas as pd
import random
from bs4 import BeautifulSoup, SoupStrainer
driver = webdriver.Chrome()
parentlist = []
linklist = []

def recursePages (htmlsource, parent): 
    for link in BeautifulSoup(htmlsource, parse_only=SoupStrainer('a')):
        if link.has_attr('href'):
            print ("----")
            mylink = link['href']
            print(link['href'])
            if ("https://online" in mylink):
                driver.get (mylink)
#                 time.sleep (1)
#                 driver.save_screenshot ("./screenpics/screen"+ str(random.randint(0,10000)) +  ".png")
                print ("\n************* MAIN PAGE *************\n")
                time.sleep(1)
                driver.save_screenshot ("./screenpics/screen"+ str(random.randint(0,10000)) +  ".png")
                try:
                    driver.find_element (By.LINK_TEXT, "Course Unit Content").click()
                    time.sleep(1)
                    driver.save_screenshot ("./screenpics/screen"+ str(random.randint(0,10000)) +  ".png")
                except:
                    print ("no link")
                parentlist.append (parent)
                linklist.append (driver.current_url)
                mystuff = driver.find_element (By.TAG_NAME, "body")    
                soup = BeautifulSoup(mystuff.get_attribute("innerHTML"), 'html.parser')
                recursePages (mystuff.get_attribute('innerHTML'), mylink)

driver.get("https://online.manchester.ac.uk/")
driver.set_window_size(1552, 840)


driver.find_element(By.NAME, "submit").click()

input ("continue?")

mystuff = driver.find_element (By.TAG_NAME, "body")

driver.save_screenshot("myscreen.png")
soup = BeautifulSoup(mystuff.get_attribute("innerHTML"), 'html.parser')
recursePages (mystuff.get_attribute('innerHTML'), "home")

mydf = pd.DataFrame ()
mydf["parent"] = parentlist
mydf["link"] = linklist

mydf
mydf.to_excel ("linklist.xlsx", index=False)
```
